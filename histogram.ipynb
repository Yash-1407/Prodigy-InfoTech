```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {'Age': [25, 30, 22, 35, 40, 28, 32, 26, 38, 29],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']}

df = pd.DataFrame(data)

# Separate data by gender
male_data = df[df['Gender'] == 'Male']['Age']
female_data = df[df['Gender'] == 'Female']['Age']

# Plotting a histogram
plt.hist([male_data, female_data], bins=10, alpha=0.5, label=['Male', 'Female'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.title('Distribution of Ages by Gender')
plt.show()
```
